```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.array([67,89,45,67,80,35,61,57])
```

```python
plt.scatter(x,range(8))
```

```python
plt.scatter(x,range(8))
plt.scatter(x.mean(),3.5,color='purple')
```

```python
std=np.sqrt(np.sum((x-x.mean())**2)/len(x))
```

```python
std
```

```python
x.std()
```

Standardization

```python

def standardize(x):
    val=(x-x.mean())/x.std() #Z-score Used to get outliers
    return val
```

```python
x_new=standardize(x)
```

```python
x_new.std()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,range(8))
plt.subplot(1,2,2)
plt.scatter(x_new,range(8),color='purple')
```

```python
df=pd.read_csv('wisc_bc_data.csv')
df
```

```python
#Separate the i/p and o/p vars
#dropping id and diag cols

x=df.drop(['id','diagnosis'],axis=1)
y=df['diagnosis']
```

```python
x.shape
```

```python
from sklearn.feature_selection import SelectKBest,chi2
skf=SelectKBest(score_func=chi2,k=5)#k=5 top 5
```

```python
skf.fit(x,y)#ip=x,op=y
```

```python

skf.get_support()
```

```python
new_x=x.iloc[:,skf.get_support()]
new_x
```
