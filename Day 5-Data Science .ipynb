```python
#Turkey's Fences
```

```python
df.describe()
```

```python
Q1=np.percentile(df['salary'],25)
Q3=np.percentile(df['salary'],75)
```

```python
Q1,Q3
```

```python
IQR=Q3-Q1
```

```python
Upper_fence=Q3+IQR*1.5
Lower_fence=Q1-IQR*1.5
```

```python
Upper_fence,Lower_fence
```

```python
ndf=df[(df['salary']<Upper_fence)&(df['salary']>Lower_fence)]
```
